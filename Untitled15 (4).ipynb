```python
# Cell 3: Data Understanding - Column Analysis
print("="*50)
print("COLUMN DESCRIPTION")
print("="*50)

# Rename columns for better understanding (if needed)
# Typical real estate dataset columns:
# X1 transaction date, X2 house age, X3 distance to MRT,
# X4 number of stores, X5 latitude, X6 longitude, Y house price

print("\nOriginal Column Names:")
for i, col in enumerate(df.columns):
    print(f"{i}: {col}")

# Check for any unnecessary columns like 'No'
if 'No' in df.columns:
    df = df.drop('No', axis=1)
    print("\nDropped 'No' column as it's just an index")

# Identify target variable (usually last column or 'Y house price of unit area')
target_col = df.columns[-1]
feature_cols = df.columns[:-1]

print(f"\nTarget Variable: {target_col}")
print(f"Feature Variables: {list(feature_cols)}")
```

```python
# Cell 5: Data Preprocessing
print("="*50)
print("DATA PREPROCESSING")
print("="*50)

# Check for outliers using IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers)

print("\nOutlier Detection:")
for col in feature_cols:
    n_outliers = detect_outliers(df, col)
    print(f"{col}: {n_outliers} outliers")

# Check for skewness
print("\nSkewness of features:")
for col in df.columns:
    skewness = df[col].skew()
    print(f"{col}: {skewness:.3f}")

# Handle outliers if necessary (optional - capping)
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower_bound, upper_bound)
    return df

# Apply capping to reduce outlier impact
for col in feature_cols:
    df = cap_outliers(df, col)

print("\nOutliers have been capped!")
```
