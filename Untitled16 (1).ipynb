```python
# ============================================
# COMPLETE LINEAR REGRESSION ASSIGNMENT
# With Real-World Data Analysis Practices
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# ============================================
# PART 1: DATA LOADING & INITIAL UNDERSTANDING
# ============================================

print("="*60)
print("STUDENT PERFORMANCE ANALYSIS - LINEAR REGRESSION")
print("="*60)

# Load dataset
df = pd.read_csv('student_performance.csv')

print("\n📊 STEP 1: DATASET LOADING & UNDERSTANDING")
print("-"*40)
print(f"Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"\nFirst 5 rows:")
print(df.head())

print(f"\nColumn Information:")
print(df.info())

print(f"\nBasic Statistics:")
print(df.describe())

# ============================================
# PART 2: EXPLORATORY DATA ANALYSIS (EDA)
# ============================================

print("\n📈 STEP 2: EXPLORATORY DATA ANALYSIS")
print("-"*40)

# 2.1 Identify Important Features (Feature Selection)
print("\n2.1 Identifying Important Features:")

# Correlation analysis to find important features
if 'Performance_Index' in df.columns:
    correlations = df.corr()['Performance_Index'].sort_values(ascending=False)
    print("\nFeature correlations with Performance Index:")
    print(correlations)

    # Plot correlation heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Feature Correlation Matrix')
    plt.tight_layout()
    plt.show()

# 2.2 Understand Data Distribution
print("\n2.2 Data Distribution Analysis:")

# Plot distributions of numerical features
numerical_cols = df.select_dtypes(include=[np.number]).columns
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols[:6]):  # Plot first 6 numerical columns
    axes[idx].hist(df[col].dropna(), bins=20, edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'Distribution of {col}')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Frequency')
    axes[idx].axvline(df[col].mean(), color='red', linestyle='--', label=f'Mean: {df[col].mean():.2f}')
    axes[idx].axvline(df[col].median(), color='green', linestyle='--', label=f'Median: {df[col].median():.2f}')
    axes[idx].legend()

plt.tight_layout()
plt.show()

# ============================================
# PART 3: DATA PREPROCESSING (Real-World Approach)
# ============================================

print("\n🔧 STEP 3: DATA PREPROCESSING")
print("-"*40)

# 3.1 Handle Missing Values
print("\n3.1 Handling Missing Values:")
missing_values = df.isnull().sum()
if missing_values.sum() > 0:
    print(f"Missing values found:\n{missing_values[missing_values > 0]}")

    # Strategy: Numerical - median, Categorical - mode
    for col in df.columns:
        if df[col].isnull().sum() > 0:
            if df[col].dtype in ['int64', 'float64']:
                df[col].fillna(df[col].median(), inplace=True)
                print(f"  - Filled missing in {col} with median: {df[col].median()}")
            else:
                df[col].fillna(df[col].mode()[0], inplace=True)
                print(f"  - Filled missing in {col} with mode: {df[col].mode()[0]}")
else:
    print("✓ No missing values found")

# 3.2 Handle Outliers using IQR method
print("\n3.2 Handling Outliers (IQR Method):")

outlier_count = 0
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    if len(outliers) > 0:
        print(f"  - {col}: {len(outliers)} outliers detected")
        # Cap outliers instead of removing
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
        outlier_count += len(outliers)

if outlier_count == 0:
    print("✓ No outliers detected")
else:
    print(f"✓ Outliers capped for {outlier_count} total instances")

# 3.3 Feature Scaling (Important for some algorithms)
print("\n3.3 Feature Scaling Analysis:")

# Separate features and target
if 'Performance_Index' in df.columns:
    target_col = 'Performance_Index'
    X = df.drop(target_col, axis=1)
    y = df[target_col]

    # Handle categorical variables
    categorical_cols = X.select_dtypes(include=['object']).columns
    if len(categorical_cols) > 0:
        print(f"Categorical columns found: {categorical_cols.tolist()}")
        print("Applying One-Hot Encoding...")
        X = pd.get_dummies(X, drop_first=True)
        print("✓ Categorical variables encoded")

    # Check if scaling is needed
    numerical_features = X.select_dtypes(include=[np.number]).columns
    print(f"\nNumerical features range check:")
    for col in numerical_features[:5]:  # Show first 5
        print(f"  {col}: min={X[col].min():.2f}, max={X[col].max():.2f}, std={X[col].std():.2f}")

    # Apply Standard Scaling
    scaler = StandardScaler()
    X_scaled = X.copy()
    X_scaled[numerical_features] = scaler.fit_transform(X[numerical_features])
    print("\n✓ Feature scaling applied (StandardScaler)")

# ============================================
# PART 4: TRAIN-TEST SPLIT & MODEL TRAINING
# ============================================

print("\n🤖 STEP 4: LINEAR REGRESSION MODEL TRAINING")
print("-"*40)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")

# Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

print("\nModel trained successfully!")
print(f"Intercept (β₀): {model.intercept_:.4f}")
print("\nTop 5 Important Features (Coefficients):")
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
}).sort_values('Coefficient', key=abs, ascending=False)
print(feature_importance.head(5).to_string(index=False))

# ============================================
# PART 5: PREDICTIONS & EVALUATION
# ============================================

print("\n📊 STEP 5: PREDICTIONS & MODEL EVALUATION")
print("-"*40)

# Make predictions
y_pred = model.predict(X_test)

# Create predictions dataframe
predictions_df = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred,
    'Residual': y_test.values - y_pred,
    'Absolute Error': np.abs(y_test.values - y_pred)
})

print("\nFirst 10 Predictions:")
print(predictions_df.head(10).to_string(index=False))

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n📈 Model Performance Metrics:")
print(f"  • Mean Squared Error (MSE): {mse:.4f}")
print(f"  • Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"  • Mean Absolute Error (MAE): {mae:.4f}")
print(f"  • R-squared (R²): {r2:.4f}")

# Performance interpretation
if r2 >= 0.8:
    print("\n✅ EXCELLENT: Model explains >80% variance in student performance")
elif r2 >= 0.6:
    print("\n👍 GOOD: Model explains 60-80% variance in student performance")
elif r2 >= 0.4:
    print("\n📊 MODERATE: Model explains 40-60% variance in student performance")
else:
    print("\n⚠️ NEEDS IMPROVEMENT: Model explains <40% variance")

# ============================================
# PART 6: INSIGHTS & VISUALIZATION
# ============================================

print("\n💡 STEP 6: KEY INSIGHTS & VISUALIZATION")
print("-"*40)

# 6.1 Actual vs Predicted Plot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Actual vs Predicted
axes[0, 0].scatter(y_test, y_pred, alpha=0.6, edgecolors='black', linewidth=0.5)
axes[0, 0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
axes[0, 0].set_xlabel('Actual Performance')
axes[0, 0].set_ylabel('Predicted Performance')
axes[0, 0].set_title('Actual vs Predicted Performance')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2: Residual Distribution
residuals = y_test - y_pred
axes[0, 1].hist(residuals, bins=20, edgecolor='black', alpha=0.7)
axes[0, 1].axvline(0, color='red', linestyle='--', linewidth=2)
axes[0, 1].set_xlabel('Residuals')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].set_title('Residual Distribution (Should be normal)')
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Feature Importance
top_features = feature_importance.head(10)
axes[1, 0].barh(top_features['Feature'], top_features['Coefficient'])
axes[1, 0].set_xlabel('Coefficient Magnitude')
axes[1, 0].set_title('Top 10 Important Features')
axes[1, 0].grid(True, alpha=0.3)

# Plot 4: Residuals vs Predicted
axes[1, 1].scatter(y_pred, residuals, alpha=0.6, edgecolors='black', linewidth=0.5)
axes[1, 1].axhline(y=0, color='red', linestyle='--', linewidth=2)
axes[1, 1].set_xlabel('Predicted Values')
axes[1, 1].set_ylabel('Residuals')
axes[1, 1].set_title('Residuals vs Predicted (Homoscedasticity check)')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================
# PART 7: REAL-WORLD INSIGHTS
# ============================================

print("\n🎯 STEP 7: BUSINESS INSIGHTS & RECOMMENDATIONS")
print("-"*40)

print("\nKey Findings from Analysis:")
print("1. Feature Importance Analysis shows which factors most affect student performance")
print("2. R² score indicates how well we can predict performance")
print("3. Residual analysis shows model reliability")

# Identify top positive and negative factors
positive_factors = feature_importance[feature_importance['Coefficient'] > 0].head(3)
negative_factors = feature_importance[feature_importance['Coefficient'] < 0].head(3)

if len(positive_factors) > 0:
    print("\n📈 Factors that POSITIVELY impact performance:")
    for _, row in positive_factors.iterrows():
        print(f"   • {row['Feature']}: +{row['Coefficient']:.4f}")

if len(negative_factors) > 0:
    print("\n📉 Factors that NEGATIVELY impact performance:")
    for _, row in negative_factors.iterrows():
        print(f"   • {row['Feature']}: {row['Coefficient']:.4f}")

print("\n💡 Recommendations for Students:")
print("• Focus on improving features with high positive coefficients")
print("• Monitor features with negative impact")
print("• Use model for early intervention strategies")

# ============================================
# SAMPLE PREDICTION FOR NEW STUDENT
# ============================================

print("\n🔮 STEP 8: PREDICT FOR NEW STUDENT")
print("-"*40)

# Create a sample student (using average values)
sample_student = X_scaled.mean().values.reshape(1, -1)
predicted_performance = model.predict(sample_student)

print(f"Predicted Performance Index for an average student: {predicted_performance[0]:.2f}")
print("\n✓ Model is ready for real-world predictions!")

print("\n" + "="*60)
print("ASSIGNMENT COMPLETED SUCCESSFULLY!")
print("All real-world data analysis practices implemented!")
print("="*60)
```
