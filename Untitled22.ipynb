```python
# Smart Quiz Performance Analyzer
# Features: Menu-driven, MCQs, Negative Marking, Difficulty Levels, Multiple Attempts Tracking, Performance Analysis

import time

# Quiz questions: (question, options, correct_option, difficulty, marks)
quiz_questions = [
    ("What is the capital of France?", ["Berlin", "Madrid", "Paris", "Lisbon"], 3, "easy", 10),
    ("Which planet is known as the Red Planet?", ["Earth", "Mars", "Jupiter", "Saturn"], 2, "easy", 10),
    ("What is 15 * 4?", ["45", "50", "60", "70"], 3, "easy", 10),
    ("Who wrote 'Romeo and Juliet'?", ["Charles Dickens", "Jane Austen", "Mark Twain", "William Shakespeare"], 4, "medium", 15),
    ("What is the square root of 64?", ["6", "7", "8", "9"], 3, "medium", 15),
    ("Which gas is most abundant in Earth's atmosphere?", ["Oxygen", "Carbon Dioxide", "Nitrogen", "Argon"], 3, "medium", 15),
    ("What is the chemical symbol for Gold?", ["Go", "Gd", "Au", "Ag"], 3, "hard", 20),
    ("Who painted the Mona Lisa?", ["Van Gogh", "Picasso", "Da Vinci", "Rembrandt"], 3, "hard", 20),
    ("What is the fastest animal on land?", ["Lion", "Cheetah", "Leopard", "Tiger"], 2, "hard", 20),
    ("Which programming language is this quiz written in?", ["Java", "C++", "Python", "Ruby"], 3, "hard", 20),
]

# Store multiple attempts: list of (score, total_possible, date_time)
attempt_history = []

def display_menu():
    print("\n" + "="*40)
    print("       SMART QUIZ PERFORMANCE ANALYZER")
    print("="*40)
    print("1. Start Quiz")
    print("2. View Last Score")
    print("3. Performance Analysis (All Attempts)")
    print("4. Leaderboard (Top Scores)")
    print("5. Exit")
    print("="*40)

def get_user_choice():
    while True:
        try:
            choice = int(input("Enter your choice (1-5): "))
            if 1 <= choice <= 5:
                return choice
            else:
                print("❌ Invalid choice. Please enter between 1 and 5.")
        except ValueError:
            print("❌ Invalid input. Please enter a number.")

def start_quiz():
    print("\n🎯 Select Difficulty Level:")
    print("1. Easy")
    print("2. Medium")
    print("3. Hard")
    print("4. Mixed (All levels)")

    diff_choice = get_user_choice()  # reuse but limit to 1-4
    while diff_choice not in [1,2,3,4]:
        print("❌ Choose 1,2,3, or 4.")
        diff_choice = get_user_choice()

    # Filter questions based on difficulty
    if diff_choice == 1:
        filtered_qs = [q for q in quiz_questions if q[4] == "easy"]
        diff_name = "Easy"
    elif diff_choice == 2:
        filtered_qs = [q for q in quiz_questions if q[4] == "medium"]
        diff_name = "Medium"
    elif diff_choice == 3:
        filtered_qs = [q for q in quiz_questions if q[4] == "hard"]
        diff_name = "Hard"
    else:
        filtered_qs = quiz_questions[:]
        diff_name = "Mixed"

    if not filtered_qs:
        print("No questions available for this difficulty.")
        return

    print(f"\n✅ Starting {diff_name} Quiz. Each correct answer gives marks, wrong answer gives NEGATIVE marks (-50% of marks).")
    print("Type 'skip' to move to next question without penalty.\n")

    score = 0
    total_possible = 0

    for idx, q in enumerate(filtered_qs, 1):
        question_text, options, correct_idx, difficulty, marks = q
        total_possible += marks

        print(f"Q{idx} [{difficulty.upper()} | +{marks} | -{marks//2} if wrong]: {question_text}")
        for opt_idx, opt in enumerate(options, 1):
            print(f"   {opt_idx}. {opt}")

        # Input validation with retry and skip
        attempt = 0
        user_answer = None
        while attempt < 2:
            ans_input = input("Your answer (1-4 or 'skip'): ").strip().lower()
            if ans_input == 'skip':
                user_answer = 'skip'
                break
            if ans_input.isdigit() and 1 <= int(ans_input) <= 4:
                user_answer = int(ans_input)
                break
            else:
                print("⚠️ Invalid input. Please enter 1-4 or 'skip'. Attempts left:", 1-attempt)
                attempt += 1

        if user_answer == 'skip':
            print("⏩ Skipped. No marks added/deducted.\n")
            continue
        elif user_answer is None:
            print("❌ No valid input. Moving to next question.\n")
            continue

        # Validate answer
        if user_answer == correct_idx:
            print(f"✅ Correct! +{marks} marks.\n")
            score += marks
        else:
            negative_marks = marks // 2  # Negative marking (custom feature)
            print(f"❌ Wrong! Correct answer was {correct_idx}. {negative_marks} marks deducted.\n")
            score -= negative_marks

    print("\n" + "="*30)
    print(f"🎉 Quiz Finished! Your Score: {score} / {total_possible}")

    # Performance classification
    percentage = (score / total_possible) * 100 if total_possible > 0 else 0
    if percentage >= 75:
        level = "Advanced 🚀"
        feedback = "Excellent! You have strong knowledge."
    elif percentage >= 45:
        level = "Intermediate 📘"
        feedback = "Good! Keep practicing to reach Advanced."
    else:
        level = "Beginner 📖"
        feedback = "Keep learning! Try again to improve."

    print(f"Performance: {level}")
    print(f"Feedback: {feedback}")

    # Store attempt
    timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
    attempt_history.append((score, total_possible, timestamp, level, percentage))

    return score, total_possible

def view_last_score():
    if not attempt_history:
        print("\n📭 No quiz attempts yet. Please start a quiz first.")
        return
    last = attempt_history[-1]
    print("\n📊 Last Quiz Score:")
    print(f"   Score: {last[0]} / {last[1]}")
    print(f"   Percentage: {last[4]:.2f}%")
    print(f"   Performance: {last[3]}")
    print(f"   Date & Time: {last[2]}")

def performance_analysis():
    if not attempt_history:
        print("\n📭 No attempts found. Start a quiz to see analysis.")
        return

    print("\n📈 PERFORMANCE ANALYSIS (All Attempts)")
    print("-"*50)
    total_attempts = len(attempt_history)
    print(f"Total Quizzes Taken: {total_attempts}")

    best_percentage = 0
    best_score = 0
    total_percentage_sum = 0

    for i, (score, total, timestamp, level, percent) in enumerate(attempt_history, 1):
        print(f"\nAttempt {i}:")
        print(f"   Score: {score}/{total} ({percent:.2f}%)")
        print(f"   Level: {level}")
        print(f"   Time: {timestamp}")

        if percent > best_percentage:
            best_percentage = percent
            best_score = score

    avg_percentage = total_percentage_sum / total_attempts if total_attempts > 0 else 0
    print("\n" + "-"*50)
    print(f"🏆 Best Performance: {best_percentage:.2f}%")
    print(f"📊 Average Performance: {avg_percentage:.2f}%")

    # Progress tracking bonus
    if total_attempts >= 2:
        last_percent = attempt_history[-1][4]
        first_percent = attempt_history[0][4]
        if last_percent > first_percent:
            print("📈 Great! Your performance is improving over time.")
        elif last_percent < first_percent:
            print("📉 Keep practicing! You can do better.")
        else:
            print("📊 Your performance is consistent.")

def leaderboard():
    if not attempt_history:
        print("\n📭 No attempts yet. Leaderboard is empty.")
        return

    # Sort by percentage (highest first)
    sorted_attempts = sorted(attempt_history, key=lambda x: x[4], reverse=True)
    print("\n🏆 LEADERBOARD (Top Scores)")
    print("-"*40)
    for rank, (score, total, timestamp, level, percent) in enumerate(sorted_attempts[:5], 1):
        print(f"{rank}. {percent:.2f}% - {level} ({score}/{total}) - {timestamp}")

# Main program loop
def main():
    while True:
        display_menu()
        choice = get_user_choice()

        if choice == 1:
            start_quiz()
        elif choice == 2:
            view_last_score()
        elif choice == 3:
            performance_analysis()
        elif choice == 4:
            leaderboard()
        elif choice == 5:
            print("\n👋 Thanks for using Smart Quiz Analyzer. Goodbye!")
            break

if __name__ == "__main__":
    main()
```
